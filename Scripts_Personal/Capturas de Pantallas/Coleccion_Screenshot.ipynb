```python

import os
import pyautogui
import sqlite3
from datetime import datetime

def capturar_guardar(ruta, nombre):
    # Verifica si la carpeta existe; si no, la crea
    if not os.path.exists(ruta):
        os.makedirs(ruta)
        print(f"Se ha creado la carpeta: {ruta}")

    # Usa pyautogui para capturar la pantalla
    imagen = pyautogui.screenshot()
    ruta_completa = os.path.join(ruta, f"{nombre}.png")
    imagen.save(ruta_completa)
    print(f"Imagen guardada en: {ruta_completa}")

# Ejemplo de uso:
ruta = "C:\\Users\\vikto\\VsCode_Screen_Shots"
nombre = "Danny Mariposa"
capturar_guardar(ruta, nombre)

# Configurar base de datos
def create_table():
    db = sqlite3.connect("navegacion.db")
    cursor = db.cursor()
    cursor.execute('''CREATE TABLE IF NOT EXISTS navegacion
                     (id INTEGER PRIMARY KEY AUTOINCREMENT,
                      url TEXT NOT NULL,
                      timestamp DATETIME NOT NULL)''') # Instruccion en SQL que cambia el nombre de la tabla y los campos según tus necesidades
    db.commit()
    db.close()

def salva_navegacion(url):
    try:
        db = sqlite3.connect("navegacion.db")
        cursor = db.cursor()
        timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        cursor.execute("INSERT INTO navegacion (url, timestamp) VALUES (?, ?)", (url, timestamp))
        db.commit()
        print("\n")
        print(f"Salvada la navegación: {url}")
    except Exception as e:
        print(f"Error saving navigation: {e}")
    finally:
        db.close()

# Inicializar la base de datos
create_table()

# Ejemplo de uso
salva_navegacion("https://www.ejemplo.com")
print("\n")
# capture_social_media() function is not defined, so this line is removed or commented out.
# capture_social_media()
```

6 - Captura de Imagenes con Python


```python
import os
import pyautogui
import random
import getpass
import shutil
import subprocess
from wand.image import Image

# Ruta donde se guardará la captura
usuario = getpass.getuser()
numaleatorio = random.randint(1, 1000)
ruta_captura = f"/mnt/c/Capturas/captura_{usuario}_{numaleatorio}.png"

# Verifica si la carpeta existe; si no, la crea
os.makedirs(os.path.dirname(ruta_captura), exist_ok=True)

# Función para verificar si un comando está disponible en el sistema
def comando_disponible(comando):
    return shutil.which(comando) is not None

# Intento de captura con PyAutoGUI
try:
    captura = pyautogui.screenshot()
    captura.save(ruta_captura)
    print(f"¡Captura guardada exitosamente con PyAutoGUI en: {ruta_captura}!")
except Exception as e:
    print(f"Error con PyAutoGUI: {e}. Probando métodos alternativos...")

    # Alternativa con Wand (ImageMagick en Python)
    try:
        with Image(width=1920, height=1080) as img:
            img.screenshot()
            img.save(filename=ruta_captura)
            print(f"¡Captura guardada con Wand (ImageMagick) en: {ruta_captura}!")
    except Exception as e:
        print(f"Error con Wand: {e}. Verifica que ImageMagick esté instalado con 'sudo apt install imagemagick'.")

    # Alternativa con xwd + convert
    if comando_disponible("xwd"):
        try:
            ruta_xwd = ruta_captura.replace(".png", ".xwd")
            subprocess.run(["xwd", "-root", "-out", ruta_xwd], check=True)
            if comando_disponible("convert"):
                subprocess.run(["convert", ruta_xwd, ruta_captura], check=True)
                print(f"¡Captura guardada con xwd en: {ruta_captura}!")
                os.remove(ruta_xwd)  # Limpieza de archivo temporal
            else:
                print("Error: ImageMagick (convert) no está disponible para convertir el archivo xwd.")
        except subprocess.CalledProcessError as e:
            print(f"Error al ejecutar xwd: {e}.")
    else:
        print("xwd no está instalado. Puedes instalarlo con: sudo apt install x11-apps")

print("Proceso finalizado.")
```

8 - Captura de imagenes

```python
import os
from shutil import move
import time
import pyautogui
#import pyscreeze

def capture_screen():
    try:
        # Capturar pantalla
        image = pyautogui.screenshot()
        # Guardar la captura con un nombre basado en el timestamp
        timestamp = str(time.time())
        filename = f"captura_{timestamp}.png"
        image.save(filename)
        print(f"Captura guardada como: {filename}")
        return filename
    except Exception as e:
        print(f"Error al capturar pantalla: {e}")
        return None

def create_folder(path):
    if not os.path.exists(path):
        os.makedirs(path)
        print(f"Folder created: {path}")

def move_file(src, dst):
    try:
        move(src, dst)
        print(f"File moved: {src} -> {dst}")
    except FileNotFoundError:
        print(f"Error: The file {src} does not exist")
    except Exception as e:
        print(f"An error occurred: {e}")

# Configurar rutas
destination_folder = "C:\\Capturas.png"

# Capturar pantalla
captura = capture_screen()

if captura:
    # Mover archivo a la carpeta de destino
    move_file(captura, destination_folder)
    
    # Generar nombre unico basado en timestamp
    timestamp = str(time.time())
    filename = f"registro_{timestamp}.txt"
    with open(filename, 'w') as f:
        f.write(f"Registro de captura: {timestamp}")
    print(f"File created: {filename}")
else:
    print("No se pudo realizar la captura de pantalla.")
```
