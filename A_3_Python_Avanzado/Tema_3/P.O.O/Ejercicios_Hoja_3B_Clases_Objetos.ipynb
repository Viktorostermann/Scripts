# CUENTA BANCARIA:

```python
'''1. Crea una clase "CuentaBancaria" con atributos como número de cuenta y
saldo. Implementa métodos para depositar y retirar dinero, y muestra el
saldo actual.'''

class CuentaBancaria:
    def __init__(self, numero_cuenta, saldo = 0):
        self.numero_cuenta = numero_cuenta
        self.saldo = saldo
        

    def depositar(self, monto):
        self.saldo += monto
        print(f"Se ha depositado {monto} EU en la cuenta {self.numero_cuenta}. Saldo actual: {self.saldo} EU")

    def retirar(self, monto):
        if self.saldo >= monto:
            self.saldo -= monto
            print(f"Se han retirado {monto} EU de la cuenta {self.numero_cuenta}. Saldo actual: {self.saldo} EU")

        else:
            print("Saldo insuficiente")

    def mostrar_saldo(self):
        print(f"El saldo actual de la cuenta {self.numero_cuenta} es de {self.saldo} EU")

    def transferencia(self, otra_cuenta, monto):
        if self.saldo >= monto:
            self.retirar(monto)
            otra_cuenta.depositar(monto)
            print(f"Se ha transferido {monto} EU desde la cuenta {self.numero_cuenta} a la cuenta {otra_cuenta.numero_cuenta}")
        else:
            print("Saldo insuficiente")

# Ejemplo de uso
cuenta1 = CuentaBancaria(123456, 500)
cuenta2 = CuentaBancaria(789012, 1000)

cuenta1.mostrar_saldo()
cuenta2.mostrar_saldo()

cuenta1.transferencia(cuenta2, 200)
cuenta1.mostrar_saldo()
cuenta2.mostrar_saldo()

cuenta1.retirar(300)
cuenta1.mostrar_saldo()

cuenta2.depositar(500)
cuenta2.mostrar_saldo()
```

# LISTA DE TAREAS

```python
'''Crea una clase "ListaTareas" que contenga una lista de tareas pendientes.
Implementa métodos para agregar una tarea, marcar una tarea como
completada y mostrar todas las tareas '''

class Lista_de_tareas:
    def __init__(self):
        self.tareas= []
    
    def agregar_tarea(self, tarea):
        self.tareas.append({"tarea": tarea, "completada": False})
        print(f"Tarea '{tarea}' agregada a la lista.")
    
    def marcar_completada(self, indice):
        if 0 <= indice < len(self.tareas):
            self.tareas[indice]["completada"] = True

    def mostrar_tareas(self):
        for i, tarea in enumerate(self.tareas):
            estado = "Completada" if tarea["completada"] else "Pendiente"
            print(f"{i + 1}. {estado}: {tarea['tarea']}")

# Ejemplo de uso
lista_tareas = Lista_de_tareas()

lista_tareas.agregar_tarea("Hacer la compra")
lista_tareas.agregar_tarea("Estudiar Python")
print()

lista_tareas.mostrar_tareas()
print()

lista_tareas.marcar_completada(0)
print()

lista_tareas.mostrar_tareas()
print()
```

# TIENDA ONLINE

```python
# TIENDA ONLINE

''' Crea una clase "Producto" con atributos como nombre, precio y cantidad en
stock. Luego, crea una clase "Tienda" que contenga una lista de productos
disponibles y métodos para agregar productos, mostrar el inventario y
realizar una compra.'''

class Producto:
    def __init__(self, nombre, precio, stock):
        self.nombre = nombre
        self.precio = precio
        self.stock = stock

class Tienda:
    def __init__(self):
        self.productos = []

    def agregar_producto(self, producto):
        self.productos.append(producto)
        print(f"Producto '{producto.nombre}' agregado a la tienda.")
    
    def mostrar_inventario(self):
        for producto in self.productos:
            print("Inventario de la tienda:")
            print(f"Nombre: {producto.nombre}, Precio: ${producto.precio}, Stock: {producto.stock}")

    def comprar_producto(self, nombre_producto, cantidad):
        for producto in self.productos:
            if producto.nombre == nombre_producto and producto.stock >= cantidad:
                total_pagar = producto.precio * cantidad
                print(f"Compra realizada: {cantidad} unidades de '{nombre_producto}'. Total a pagar: ${total_pagar}")
                producto.stock -= cantidad
                break
        else:
            print(f"No hay suficientes unidades del producto '{nombre_producto}' en stock.")

    def realizar_compra(self, nombre_producto, cantidad):
        """Alias compatibilidad: llama a comprar_producto."""
        return self.comprar_producto(nombre_producto, cantidad)

# Ejemplo de uso
tienda_online = Tienda()

print()
producto1 = Producto("Camiseta", 20.0, 50)
producto2 = Producto("Zapatos", 50.0, 30)
print()

tienda_online.agregar_producto(producto1)
print()

tienda_online.agregar_producto(producto2)
print()

tienda_online.mostrar_inventario()
print()

tienda_online.realizar_compra("Camiseta", 10)
print()

tienda_online.realizar_compra("Zapatos", 20)
print()

tienda_online.realizar_compra("Gorra", 5)
print()

tienda_online.mostrar_inventario()
print()
```

# PILA (STACK) BÁSICA

```python
''' En programación, un "stack" es una estructura de datos que sigue el
principio de LIFO (Last In, First Out), lo que signi ca que el último elemento
agregado a la pila es el primero en ser retirado. Imagina una pila de platos:
cuando apilas un nuevo plato, este se coloca en la parte superior de la pila,
y cuando retiras un plato, siempre tomas el de arriba primero.

En Python, puedes implementar un stack utilizando una lista. Puedes
agregar elementos a la pila utilizando el método `append()`, y puedes retirar
elementos de la pila utilizando el método `pop()` sin ningún índice
especi cado, ya que `pop()` por defecto elimina y devuelve el último
elemento de la lista.Los stacks son útiles en muchas situaciones, como algoritmos de búsqueda
y recorrido, manejo de llamadas a funciones (con la pila de llamadas),
manejos de historial y navegación y más.

Crea una clase "Pila" que represente una pila (stack) básica. Implementa
métodos para agregar elementos a la pila, quitar elementos y mostrar el
contenido actual.

Por supuesto, estaré encantado de explicarte qué es un "stack" en el
contexto de la programación y cómo se utiliza en Python.'''

class Stack:
    def __init__(self):
        self.items = []

    def agregar(self, item):
        self.items.append(item)

    def quitar(self):
        if not self.es_vacia():
           return self.items.pop()
        else:
            raise IndexError("La pila está vacía")

    def esta_vacia(self):
        return len(self.items) == 0

    def mostrar_pila(self): # Muestra los valores de la pila o Stack
        return self.items
    
# Ejemplo de uso

pila = Stack()

pila.push(1)
pila.push(2)
pila.push(3)

print(pila.show_stack())

print(pila.peek()) # Muestra el ultimo valor ingresado
print(pila.size()) # Muestra el tamaño de la pila
print(pila.pop()) # Elimina el ultimo valor ingresado

print(pila.show_stack())

print(pila.is_empty()) # Verifica si esta vacia o no
print(pila.peek()) # Muestra el ultimo valor ingresado
print(pila.size()) # Muestra el tamaño de la pila
```

# SISTEMA DE GESTION DE BIBLIOTECA

```python
''' Crea un sistema de gestión de una biblioteca utilizando clases en Python.
Debes implementar las siguientes clases:

1. “Libro”: Representa un libro con atributos como título, autor y número de
ejemplares disponibles.

2. “Usuario”: Representa a un usuario de la biblioteca con atributos como
nombre, número de identi cación y lista de libros prestados.

3. “Biblioteca”: Representa la biblioteca en sí, con métodos para agregar
libros, prestar libros a usuarios, devolver libros y mostrar el inventario.'''

class Libro:
    def __init__(self, titulo, autor, ejemplares_disponibles):
        self.titulo = titulo
        self.autor = autor
        self.ejemplares_disponibles = ejemplares_disponibles

class Usuario:
    def __init__(self, nombre, id_usuario, libros_prestados=[]):
        self.nombre = nombre
        self.id_usuario = id_usuario
        self.libros_prestados = []

class Biblioteca:
    def __init__(self):
        self.libros = []
        self.usuarios = []

    def agregar_libro(self, libro):
        self.libros.append(libro)

    def prestar_libro(self, usuario_id, libro_titulo):
        usuario = next((u for u in self.usuarios if u.id_usuario == usuario_id), None)
        libro = next((l for l in self.libros if l.titulo == libro_titulo), None)

        if usuario is None or libro is None:
            print("Usuario o libro no encontrado.")
            print()
            return

        if libro.ejemplares_disponibles > 0:
            libro.ejemplares_disponibles -= 1
            usuario.libros_prestados.append(libro)
            print(f"Libro '{libro.titulo}' prestado a {usuario.nombre}.")
            print()
        else:
            print("No hay ejemplares disponibles.")
            print()
            return

        if libro.ejemplares_disponibles == 0:
            self.libros.remove(libro)
        else:
            print("No se pudo encontrar el libro en la biblioteca.")
            print()
            return

    def devolver_libro(self, usuario_id, libro_titulo):
        usuario = next((u for u in self.usuarios if u.id_usuario == usuario_id), None)
        libro = next((l for l in self.libros if l.titulo == libro_titulo), None)
        if usuario is None or libro is None:
            print("Usuario o libro no encontrado.")
            print()
            return
        if libro in usuario.libros_prestados:
            usuario.libros_prestados.remove(libro)
            libro.ejemplares_disponibles += 1
            print(f"Libro '{libro.titulo}' devuelto por {usuario.nombre}.")
        else:
            print()
            print("El usuario no tiene ese libro prestado.")
            print()
            return

    def mostrar_inventario(self):
        print()
        print("Inventario de la biblioteca:")
        print()
        for libro in self.libros:
            print(f"Título: {libro.titulo}, Autor: {libro.autor}, Ejemplares Disponibles: {libro.ejemplares_disponibles}")

# Ejemplo de uso
biblioteca = Biblioteca()

libro1 = Libro("Harry Potter y la Piedra Filosofal", "J.K. Rowling", 5)
libro2 = Libro("Cien años de soledad", "Gabriel García Márquez", 3)
libro3 = Libro("Don Quijote de la Mancha", "Miguel de Cervantes", 2)
biblioteca.agregar_libro(libro1)
biblioteca.agregar_libro(libro2)
biblioteca.agregar_libro(libro3)

usuario1 = Usuario("Juan Pérez", 1)
usuario2 = Usuario("María Rodríguez", 2)
biblioteca.usuarios.append(usuario1)
biblioteca.usuarios.append(usuario2)

biblioteca.prestar_libro(1, "Harry Potter y la Piedra Filosofal")
biblioteca.prestar_libro(2, "Cien años de soledad")
biblioteca.prestar_libro(1, "Don Quijote de la Mancha")

biblioteca.devolver_libro(1, "Harry Potter y la Piedra Filosofal")

biblioteca.mostrar_inventario()

```
