```python
# EJERCICIO CLASE PERSONA

''' Define una clase Persona con atributos como nombre, edad y profesión.
Imprime un método que permita a la persona presentarse a sí misma. '''


class Persona:
    def __init__(self, nombre, edad, profesion):
        self.nombre = nombre
        self.edad = edad
        self.profesion = profesion

    def presentarse(self):
        return f"Hola, mi nombre es {self.nombre}, tengo {self.edad} años y soy {self.profesion}."

# EJEMPLO DE USO
persona1 = Persona("Ana", 30, "Ingeniera")
print(persona1.presentarse())
```

```python
# EJERCICIO CALCULADORA BÁSICA

''' Crea una clase llamada “Calculadora” con métodos para sumar, restar, multiplicar y dividir. 
Crea objetos de esta clase y realiza algunas operaciones básicas. Imprime los resultados. '''

class Calculadora:
    def sumar(self, a, b):
        return a+b
    
    def restar(self, a, b):
        return a-b
    
    def multiplicar(self, a, b):
        return a*b
    
    def dividir(self, a, b):

        if b != 0:
            return a/b
        else:
            return "Error: División por cero"

# EJEMPLO DE USO

calc = Calculadora()
print("Suma:", calc.sumar(10, 5))
print("Resta:", calc.restar(10, 5))
print("Multiplicación:", calc.multiplicar(10, 5))
print("División:", calc.dividir(10, 5))
print("División por cero:", calc.dividir(10, 0))

    

```

```python
# EJERCICIO LIBRO

''' Crea una clase “Libro” con atributos como título, autor y año de publicación. 
Luego, crea varios objetos Libro y muestra su información.'''

class Libro:
    def __init__(self, titulo, autor, anio_publicacion):
        self.titulo = titulo
        self.autor = autor
        self.anio_publicacion = anio_publicacion

    def mostrar_info(self):
        return f"'{self.titulo}' por {self.autor}, publicado en {self.anio_publicacion}."

# EJEMPLO DE USO

libro1 = Libro("Cien Años de Soledad", "Gabriel García Márquez", 1967)
libro2 = Libro("1984", "George Orwell", 1949)

print(libro1.mostrar_info())
print(libro2.mostrar_info())






```

```python
# EJERCICIO RECTÁNGULO

''' Crea una clase “Rectangulo” con atributos de longitud y ancho.
Implementa un método para calcular el área y el perímetro del rectángulo. '''


class Rectangulo:
    def __init__(self, longitud, ancho):
        self.longitud = longitud
        self.ancho = ancho

    def area(self):
        return self.longitud * self.ancho

    def perimetro(self):
        return 2 * (self.longitud + self.ancho) 
    
# EJEMPLO DE USO
rectangulo1 = Rectangulo(5, 3)
print("Área del rectángulo:", rectangulo1.area())
print("Perímetro del rectángulo:", rectangulo1.perimetro())
```

```python

# EJERCICIO CÍRCULO
''' Crea una clase “Círculo” con un atributo de radio.
Implementa métodos para calcular el área y la circunferencia del círculo. '''
import math

class Circulo:
    def __init__(self, radio):
        self.radio = radio

    def area(self):
        return math.pi * (self.radio ** 2)

    def circunferencia(self):
        return 2 * math.pi * self.radio

# EJEMPLO DE USO
circulo1 = Circulo(4)
print("Área del círculo:", circulo1.area())
print("Circunferencia del círculo:", circulo1.circunferencia())
```

```python
# EJERCICIO BANCO

''' Crea una clase “CuentaBancaria” con atributos como titular, saldo y número de cuenta.
Implementa métodos para depositar, retirar y mostrar el saldo actual. '''

class CuentaBancaria:

    def __init__(self, titular, numero_cuenta, saldo=0):
        self.titular = titular
        self.numero_cuenta = numero_cuenta
        self.saldo = saldo
    def depositar(self, cantidad):
        self.saldo += cantidad
        return f"Depósito de {cantidad} realizado. Saldo actual: {self.saldo}"
    def retirar(self, cantidad):
        if cantidad <= self.saldo:
            self.saldo -= cantidad
            return f"Retiro de {cantidad} realizado. Saldo actual: {self.saldo}"
        else:
            return "Fondos insuficientes"
    def mostrar_saldo(self):
        return f"Saldo actual: {self.saldo}"
    
# EJEMPLO DE USO
cuenta1 = CuentaBancaria("Carlos Pérez", "123456789")
print(cuenta1.depositar(500))
print(cuenta1.retirar(200))
print(cuenta1.mostrar_saldo())  
```

```python


# EJERCICIO COCHE

''' Crea una clase “Coche” con atributos como marca, modelo y año. 
Implementa un método para encender el coche y otro para apagarlo 
(puedes simulae el encendido y apagado con una variable booleana).'''

class Coche:
    def __init__(self, marca, modelo, anio):
        self.marca = marca
        self.modelo = modelo
        self.anio = anio
        self.encendido = False

    def encender(self):
        if not self.encendido:
            self.encendido = True
            return "El coche está encendido."
        else:
            return "El coche ya está encendido."

    def apagar(self):
        if self.encendido:
            self.encendido = False
            return "El coche está apagado."
        else:
            return "El coche ya está apagado."

# EJEMPLO DE USO
coche1 = Coche("Toyota", "Corolla", 2020)
print(coche1.encender())
print(coche1.apagar())
```
