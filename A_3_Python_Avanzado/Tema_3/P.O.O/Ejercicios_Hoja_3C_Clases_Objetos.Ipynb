# SISTEMA DE RESERVAS DE VUELOS:

Imagina que estás desarrollando un sistema de reservas de vuelos para una
aerolínea. Crea un sistema de clases que permita a los usuarios realizar
reservas de vuelos. Aquí tienes una posible estructura:

- Clase base: `Vuelo`
- Atributos: número de vuelo, origen, destino, capacidad máxima, lista de
pasajeros
- Métodos: agregar pasajero, veri car disponibilidad de asientos
- Clase derivada: `VueloEspecial` (hereda de `Vuelo`)
- Atributos adicionales: motivo del vuelo especial (por ejemplo, vacaciones,
trabajo)

Resuelve el problema creando instancias de estas clases y realizando
reservas para diferentes vuelos y tipos de vuelos especiales.

```python
# SISTEMA DE RESERVAS DE VUELOS:

''' Imagina que estás desarrollando un sistema de reservas de vuelos para una
aerolínea. Crea un sistema de clases que permita a los usuarios realizar
reservas de vuelos. Aquí tienes una posible estructura:

- Clase base: `Vuelo`
- Atributos: número de vuelo, origen, destino, capacidad máxima, lista de
pasajeros
- Métodos: agregar pasajero, veri car disponibilidad de asientos
- Clase derivada: `VueloEspecial` (hereda de `Vuelo`)
- Atributos adicionales: motivo del vuelo especial (por ejemplo, vacaciones,
trabajo)

Resuelve el problema creando instancias de estas clases y realizando
reservas para diferentes vuelos y tipos de vuelos especiales.'''

class Vuelo:

    def __init__(self, numero_vuelo, origen, destino, capacidad_maxima):
        self.numero_vuelo = numero_vuelo
        self.origen = origen
        self.destino = destino
        self.capacidad_maxima = capacidad_maxima
        self.lista_pasajeros = []

    def verificar_disponibilidad_asientos(self):
        return (self.capacidad_maxima - len(self.lista_pasajeros))

    def agregar_pasajero(self, pasajero):
        if self.verificar_disponibilidad_asientos() > 0:
            self.lista_pasajeros.append(pasajero)
            print(f"Pasajero {pasajero} agregado al vuelo {self.numero_vuelo}.")
        else:
            print(f"No se pueden agregar más pasajeros al vuelo {self.numero_vuelo}. Capacidad el vuelo tiene la capacidad mmáxima alcanzada.")
            print("")

class VueloEspecial(Vuelo): # Herencia de clases (Hija de la clase Vuelo)
    def __init__(self, numero_vuelo, origen, destino, capacidad_maxima, motivo):
        super().__init__(numero_vuelo, origen, destino, capacidad_maxima) # Llama al constructor de la clase base (Herencia de clases) Vuelo.
        self.motivo = motivo



## Ejemplo de uso Vuelo Regular ----------------------------------------

# Crear instancia de Vuelo
vuelo_regular = Vuelo("ABC123", "NY", "MAD", 150)

# Realizar reservas en el vuelo normal
vuelo_regular.agregar_pasajero("Alice")
vuelo_regular.agregar_pasajero("Bob")

print("")
print(f"Vuelo Regular: Asientos disponibles en el vuelo {vuelo_regular.numero_vuelo}: {vuelo_regular.verificar_disponibilidad_asientos()}") # Asientos disponibles en el vuelo ABC123: 98
print("")

## # Ejemplo de uso Vuelo Especial --------------------------------------

# Crear instancia de Vuelo Especial
vuelo_especial = VueloEspecial("XYZ789", "MIA", "BAR", 100, "Vacaciones")

print("")
# Realizar reservas en el vuelo especial
vuelo_especial.agregar_pasajero("Charlie")
vuelo_especial.agregar_pasajero("David")

print("")
print(f"Vuelo Especial: Asientos disponibles en el vuelo {vuelo_especial.numero_vuelo}: {vuelo_especial.verificar_disponibilidad_asientos()}") # Asientos disponibles en el vuelo XYZ789: 98
print(f"Motivo del vuelo especial: {vuelo_especial.motivo}") # Motivo del vuelo especial: Vacaciones


```

''' Supongamos que estás construyendo un sistema de gestión de empleados

- Clase base: `Empleado`

- Clase derivada: `EmpleadoTiempoCompleto` (hereda de `Empleado`)

```python
# SISTEMA DE GESTIÓN DE EMPLEADOS: 

''' Supongamos que estás construyendo un sistema de gestión de empleados
para una empresa. Crea un sistema de clases que maneje la información de
los empleados, incluyendo empleados a tiempo completo y empleados a
tiempo parcial.

- Clase base: `Empleado`
- Atributos: nombre, apellido, salario base

- Clase derivada: `EmpleadoTiempoCompleto` (hereda de `Empleado`)
- Atributo adicional: bono anual

- Clase derivada: `EmpleadoTiempoParcial` (hereda de `Empleado`)
- Atributo adicional: horas trabajadas por semana. Resuelve el problema creando instancias de estas clases y calculando los
salarios totales para diferentes tipos de empleados.'''

class Empleado_base:
    def __init__(self, nombre, apellido, salario_base):
        self.nombre = nombre
        self.apellido = apellido
        self.salario_base = salario_base

    def calcular_salario_total(self):
        return self.salario_base
    
    def mostrar_info_empleado(self):
        print(f"Nombre: {self.nombre} {self.apellido}")
        print(f"Salario Base: ${self.salario_base:.2f}")

class EmpleadoTiempoExtendido(Empleado_base):
    def __init__(self, nombre, apellido, salario_base, bono_anual):
        super().__init__(nombre, apellido, salario_base)
        self.bono_anual = bono_anual

    def calcular_salario_total(self):
        return self.salario_base + self.bono_anual
    
    def mostrar_info_empleado(self):
        super().mostrar_info_empleado()
        print(f"Bono Anual: ${self.bono_anual:.2f}")
        print("Empleado a Tiempo Extendido")

class EmpleadoTiempoParcial(Empleado_base):
    def __init__(self, nombre, apellido, salario_base, horas_trabajadas_semana):
        super().__init__(nombre, apellido, salario_base)
        self.horas_trabajadas_semana = horas_trabajadas_semana

    def calcular_salario_total(self):    
        return (12*4*self.salario_base) + self.salario_base * self.horas_trabajadas_semana
    
    def mostrar_info_empleado(self):
        super().mostrar_info_empleado()
        print(f"Horas Trabajadas Semanalmente: {self.horas_trabajadas_semana}")
        print("Empleado a Tiempo Parcial")

# Ejemplo de uso

Lista1 = EmpleadoTiempoExtendido("Juan", "Pérez", 5000, 2000)

Lista2 = EmpleadoTiempoParcial("María", "García", 30, 25)

Lista3 = Empleado_base("Carlos", "López", 4500) 

empleados = [Lista1, Lista2, Lista3]

# Imprimir información de los empleados y salarios totales
for empleado in empleados:
    print("\n** Información del Empleado:")
    print("")
    empleado.mostrar_info_empleado()
    print(f"Salario Total: ${empleado.calcular_salario_total():.2f} USD _______")


```
