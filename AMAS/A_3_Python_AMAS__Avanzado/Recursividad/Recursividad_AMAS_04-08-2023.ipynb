# ¿Que es la recursividad? 

# La recursividad es una técnica de programación que permite a una función llamar a sí misma para resolver un problema complejo en partes más simples.

#Caracteristicas de la recursividad:
 1. La función debe tener una condición de parada.
 2. La función debe llamar a sí misma.
 3. La función debe tener una regla de recursividad.



```python

# Ejemplo de recursividad en Python: Factorial

def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)

print(factorial(5))
```

# Explicacion:

En la funcion anterior, el caso base n es igual a 0 o 1, en cuyo caso, se devuelve directamente (el factorial de 0 y 1 es 1)
Si n no es igual a 0 o 1, la funcion se llama a si misma con el argumento n - 1.
En cada llamada recursiva, se reduce el valor de n en uno, acercandose asi al caso base.
La mutiplicacion n * factorial (n-1) se realiza en cada llamada recursiva, y el resultado final es factorial de n.

# COnsideraciones finales:

Las funciones recursivas pueden ser poderosas, pero tambien pueden ser ineficientes en terminos de uso de memoria y tiempo de ejecucion en Problemas que pueden resolverse de manera mas eficiente con iteracion o enfoques no recursivos, es preferible utilizar esas soluciones.
